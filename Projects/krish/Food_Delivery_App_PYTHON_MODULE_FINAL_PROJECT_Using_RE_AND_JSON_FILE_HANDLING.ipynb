```python
import re
import json

class Restaurant:
    def __init__(self):
        self.Restro_Name = "The Park"
        self.food = {}
        self.food_id = len(self.food) + 1
        self.Admin_details = {}
        self.Admin_details_id = len(self.Admin_details) + 1
        self.user_details = {}
        self.user_details_id = len(self.user_details) + 2
        self.ORDERED_ITEMS = []
        print("Welcome to {} Restaurant".format(self.Restro_Name))

    def Admin_register(self):
        try:
            print("\n  Welcome to {} Restaurant".format(self.Restro_Name))
            self.Admin_email = input("Enter Admin mail id : ")
            Email_matcher = re.finditer("@", self.Admin_email)
            for i in Email_matcher:
                y = i.start()
            if self.Admin_email[y + 1:] == "hotmail.com":
                print("It is a hotmail Id")
            elif self.Admin_email[y + 1:] == "yahoo.com":
                print("It is a Yahoo Id")
            elif self.Admin_email[y + 1:] == "outlook.com":
                print("It is a outlook Id")
            elif self.Admin_email[y + 1:] == "gmail.com":
                print("It is a Gmail Id")
            else:
                print("Enter Valid mail Id")
            self.Admin_password = input("Enter Admin password : ")
            if len(self.Admin_password) >= 8:
                print("Valid Password ")
            else:
                print("Invalid Password")
            Admin_details = {"Email id": self.Admin_email, "Password": self.Admin_password}
            self.Admin_details_id = len(self.Admin_details) + 1
            self.Admin_details[self.Admin_details_id] = Admin_details
            print("\nAdmin Account is Created Successfully\n")
            with open("Restaurant_details_2.json", "w") as f:
                json.dump({"Restaurant data": [{"Admin details": self.Admin_details}]}, f)
            print("Admin Details")
            for i in self.Admin_details:
                print(i, ":", self.Admin_details[i])
        except Exception as e:
            print("\nPlease provide valid info")

    def Admin_login(self):
        try:
            print("\n $ Welcome to {} Restaurant $".format(self.Restro_Name))
            Email = input("Enter mail id : ")
            Password = input("Enter Passwrod : ")
            with open("Restaurant_details_2.json", "r") as f:
                Registered_Admin = json.loads(f.read())
                Mail_id1 = Registered_Admin["Restaurant data"][0]["Admin details"]["1"]["Email id"]
                Mail_id2 = Registered_Admin["Restaurant data"][0]["Admin details"]["2"]["Email id"]
                password1 = Registered_Admin["Restaurant data"][0]["Admin details"]["1"]["Password"]
                password2 = Registered_Admin["Restaurant data"][0]["Admin details"]["2"]["Password"]
            if Mail_id1 or Mail_id2:
                if Email == Mail_id1 or Email == Mail_id2 and Password == password1 or Password == password2:
                    print("Admin Login Successfull")
                    while True:
                        print("Select your Option")
                        print("1. Add food item\n2. Edit Food items\n3. View Food items\n4. Del Food items\n5. Exit")
                        z = input("Enter your Option : ")
                        if z == "1":
                            self.Add_food_item()
                        elif z == "2":
                            self.Edit_food_item()
                        elif z == "3":
                            self.View_food_item()
                        elif z == "4":
                            self.Del_food_item()
                        elif z == "5":
                            break
                        else:
                            print("Plese Enter Valid Option")
                else:
                    print("Please Enter valid mail or Password")
            else:
                print("Admin Not found")
        except Exception as e:
            print("\n Provide Valid Inputs")

    def Add_food_item(self):
        try:
            Name = input("Enter Food item name : ")
            quantity = float(input("Enter the quantity in pieces or kg's or grams : "))
            price = int(input("Enter the Price  in INR : "))
            stock = int(input("Enter the Stock : "))
            discount = int(input("Enter the discount in percentage : "))
            food_item = {"Name": Name, "Quantity": quantity, "Price": price, "Stock": stock, "Discount": discount}
            self.food_id = len(self.food) + 1
            self.food[self.food_id] = food_item
            print("Food item added successfully")
            with open("Restaurant_details_2.json", "w+") as f:
                json.dump({"Restaurant data": [{"Admin details": self.Admin_details}, {"Menu": self.food}]}, f)
            print("\n Newly added food items : ", self.food, " \n ")
        except Exception as e:
            print("\nPlease provide valid inputs")

    def Edit_food_item(self):
        try:
            x = str(input("Enter food id that you want to update : "))
            with open("Restaurant_details_2.json", "r") as f:
                food = json.loads(f.read())
                food_items = food['Restaurant data'][1]['Menu']
            if x in food_items.keys():
                print("1. Update food name\n2. update quantity\n3. update price\n4. update stock\n5. update discount")
                y = input("Enter the key that you want edit : ")
                if y == "1":
                    food_items[x]["Name"] = input("Enter your new food name : ")
                    print("\n Successfully Updated \n")
                elif y == "2":
                    food_items[x]["Quantity"] = float(input("Enter the updated quantity in pieces or Kg's or grams : "))
                    print("\n Successfully Updated \n")
                elif y == "3":
                    food_items[x]["Price"] = int(input("Enter the updated price : "))
                    print("\n Successfully Updated \n")
                elif y == "4":
                    food_items[x]["Stock"] = int(input("Enter the updated stock : "))
                    print("\n Successfully Updated \n")
                elif y == "5":
                    food_items[x]["Discount"] = int(input("Enter the updated discount in percentage : "))
                    print("\n Successfully Updated \n")
                else:
                    print("Invalid option")
                with open("Restaurant_details_2.json", "w+") as f:
                    json.dump({"Restaurant data": [{"Admin details": self.Admin_details}, {"Menu": food_items}]}, f)
            else:
                print("Incorrect food id")
        except Exception as e:
            print("Please provide valid inputs")

    def View_food_item(self):
        print("list of food items")
        with open("Restaurant_details_2.json", "r") as f:
            MENU = json.loads(f.read())
            print(MENU)
            VFI = MENU["Restaurant data"][1]["Menu"]
        if len(VFI.keys()) != 0:
            for i in VFI:
                print("food_id: ", {i})
                for j in VFI[i]:
                    print(j, ":", VFI[i][j])
        else:
            print("Food items not found")

    def Del_food_item(self):
        try:
            print("Enter the food id that you want to delete\n")
            x = input("Enter the food id : ")
            with open("Restaurant_details_2.json", "r") as f:
                MENU = json.loads(f.read())
                DFI = MENU["Restaurant data"][1]["Menu"]
                f.close()
                if x in DFI.keys():
                    del DFI[x]
                    print("Food item deleted successfully")
                    print("The remaining food items are : ", DFI)
                    MENU["Restaurant data"][1]["Menu"] = DFI
                    print('Menuuuuu', MENU)
                    with open("Restaurant_details_2.json", "w+") as fp:
                        json.dump(MENU, fp)
                        fp.close()
                else:
                    print("Incorrect food id")
        except Exception as e:
            print("Please provide valid input")

    def user_register(self):
        try:
            print("\n Welcome to {} Restaurant".format(self.Restro_Name))
            self.fullname = input("Enter a fullname :  ")
            self.mobile_number = input("Enter your number : ")
            m = re.fullmatch("[6-9][0-9]{9}", self.mobile_number)
            if m!=None:
                print("Valid Mobile Number")
            else:
                print("Not a Valid Mobile Number")
            with open("Restaurant_details_2.json", "r") as f:
                self.all_details = json.loads(f.read())
                f.close()
            self.mail_id = input("Enter your mail id : ")
            Email_matcher1 = re.finditer("@", self.mail_id)
            for i in Email_matcher1:
                y = i.start()
            if self.mail_id[y + 1:] == "hotmail.com":
                print("It is a hotmail Id")
            elif self.mail_id[y + 1:] == "yahoo.com":
                print("It is a Yahoo Id")
            elif self.mail_id[y + 1:] == "outlook.com":
                print("It is a outlook Id")
            elif self.mail_id[y + 1:] == "gmail.com":
                print("It is a Gmail Id")
            else:
                print("Enter Valid mail Id")
            self.password = input("Enter your password : ")
            if len(self.password) >= 8:
                print("Valid Password ")
            else:
                print("Invalid Password")
            self.address = input("Enter your Address : ")
            user_details = {"Full-Name": self.fullname, "Mobile Number": self.mobile_number, "Mail-id": self.mail_id,"Password": self.password, "Address": self.address}
            self.user_details_id = len(self.user_details) + 1
            self.user_details[self.user_details_id] = user_details
            print("\n User Account Created Successfully\n")
            print("User Details")
            print("All details", self.all_details)
            for details in self.all_details['Restaurant data']:
                self.all_details['User Details'] = self.user_details
            with open("Restaurant_details_2.json", "w+") as f:
                json.dump(self.all_details, f)
            for i in self.user_details:
                print(i, ":", self.user_details[i])
        except exception as e:
            print("\n Enter Valid inputs")

    def user_login(self):
        try:
            print("Welcome to {}  Restaurant \n".format(self.Restro_Name))
            mail_id = input("Enter mail id : ")
            password = input("Enter Passwrod : ")
            with open("Restaurant_details_2.json", "r") as f:
                Registered_User = json.loads(f.read())  
                f.close()
            user_details = Registered_User["User Details"]
            for i in user_details:
                if mail_id == user_details[i]['Mail-id']:
                    self.id = i
                else:
                    print("Incorrect Mail-id or Password")
            print("User Login Successfull")
            while True:
                print("1. Place New Order  \n2. Check Order History  \n3. Update Your Profile  \n4.Exit")
                a = input("Select Your Option :  ")
                if a == "1":
                    self.Place_Order(mail_id)
                elif a == "2":
                    self.ORDERED_HISTORY(mail_id)
                elif a == "3":
                    self.Update_Details(mail_id)
                elif a == "4":
                    break
                else:
                    print("Enter Valid option")
        except exception as e:
            print("\nInvalid Inputs")

    def Place_Order(self, mail_id):
        try:
            with open("Restaurant_details_2.json", "r") as f:
                self.MENU = json.loads(f.read())
                f.close()
            item_details = self.MENU["Restaurant data"][1]["Menu"]
            user_details = self.MENU['User Details']
            for i in user_details:
                if mail_id == user_details[i]['Mail-id']:
                    self.id = i
                    self.data = user_details[i]
                else:
                    print("No Orders to Show")
            if len(item_details.keys()) != 0:
                print("Menu Card")
                menu = []
                for items in item_details:
                    print('Items', items)
                    menu.append([item_details[items]["Name"], item_details[items]["Quantity"], item_details[items]["Price"]])
                print(menu)
                while True:
                    print("Select your option")
                    print("1. Place Order\n2. Exit")
                    b = input("Enter your option :  ")
                    if b == "1":
                        print("Select your Food Items from our Menu : ")
                        c = input().split(",")
                        for i in c:
                            d = int(i)
                            if d <= len(menu):
                                self.ORDERED_ITEMS.append(menu[d - 1])
                            else:
                                print("Out Of Stock")
                        print("Your Selected items are : ")
                        self.data['Order History'] = self.ORDERED_ITEMS
                        user_details[self.id] = self.data
                        with open("Restaurant_details_2.json", "w+") as f:
                            json.dump(self.MENU, f)
                            f.close()
                        for j in self.ORDERED_ITEMS:
                            print(j)
                    elif b == "2":
                        break
                    else:
                        print("Please Select a valid option")
            else:
                print("There is no food")
        except exception as e:
            print("Please Provide Valid inputs")

    def ORDERED_HISTORY(self, mail_id):
        with open("Restaurant_details_2.json", "r") as f:
            self.details = json.loads(f.read())
        user_details = self.details['User Details']
        for i in user_details:
            if mail_id == user_details[i]['Mail-id']:
                order_history = self.details['User Details'][i]
                if 'Order History' in order_history.keys():
                    print('Your Previous Orders are \n', order_history['Order History'])
                else:
                    print("No Orders to Show")
                break
            else:
                print("No Orders to Show")

    def Update_Details(self, mail_id):
        try:
            with open("Restaurant_details_2.json", "r") as f:
                self.all_details = json.loads(f.read())
                f.close()
            user_details = self.all_details['User Details']
            for i in user_details:
                if mail_id == user_details[i]['Mail-id']:
                    self.id = i
                else:
                    pass
            print("1. Edit Full Name  \n2. Edit Mobile Number  \n3. Edit Mail Id  \n4. Change Password  \n5. Edit Address  \n6. Exit")
            e = input("Select Your Option :  ")
            if e == "1":
                user_details[self.id]["Full-Name"] = input("Enter your Updated Full Name :  ")
                print("Your Full Name is Updated Successfully")
            elif e == "2":
                user_details[self.id]["Mobile Number"] = input("Enter Your Updated Mobile Number :  ")
                print("Your Mobile Number is Updated Successfully")
            elif e == "3":
                user_details[self.id]["Mail-id"] = input("Enter Your Updated Mail-id :  ")
                print("Your Mail-id is updated successfully")
            elif e == "4":
                user_details[self.id]["Password"] = input("Enter your new password :  ")
            elif e == "5":
                user_details[self.id]["Address"] = input("Enter your Updated Address :  ")
                print("Your Address Updated Successfully")
            else:
                print("Select valid option")
            with open("Restaurant_details_2.json", "w") as f:
                json.dump(self.all_details, f)
        except exception as e:
            print("Invalid Inputs")


def main():
    anudeep = Restaurant()
    print("\n WELCOME TO THE PARK RESTAURANT")
    while True:
        print("1. Admin  \n2. User  \n3. Exit")
        k = input("Select Your Option :  ")
        if k == "1":
            while True:
                print("Select the Below Option")
                print("1. Admin_to_Register  \n2. Admin_Login  \n3. Exit")
                l = input("Enter your option : ")
                if l == "1":
                    anudeep.Admin_register()
                elif l == "2":
                    anudeep.Admin_login()
                elif l == "3":
                    break
                else:
                    print("Please select valid input")
        elif k == "2":
            while True:
                print("1. User Register  \n2. User Login  \n3. Exit")
                m = input("Select Your Option :  ")
                if m == "1":
                    anudeep.user_register()
                elif m == "2":
                    anudeep.user_login()
                elif m == "3":
                    break
                else:
                    print("Please Select Your Valid Option")
        elif k == "3":
            break
        else:
            print("OOPS! You Selected Wrong Option PLease Select Valid option")


if __name__ == '__main__':
    main()
print("*****************************************************************************************************************")
print("THANK YOU FOR VISITING OUR RESTAURANT")
print("*****************************************************************************************************************")
```
