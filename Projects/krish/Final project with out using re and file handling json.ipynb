```python
#WITHOUT USING RE MODULE :- 
class Restaurant:
    def __init__(self):
        self.Restro_Name = "The Park"
        self.food = {}
        self.food_id = len(self.food) + 1
        self.Admin_details = {}
        self.user_details = {}
        self.ordered_item = []
        print("Welcome TO {} Restaurant".format(self.Restro_Name))
    def Admin_register(self):
        try:
            self.Admin_email = input("Enter Admin mail id : ")
            self.Admin_password = input("Enter Admin password : ")
            self.Admin_details = {"Email id" : self.Admin_email , "Password" : self.Admin_password}
            print("\nAdmin Account is Created Successfully\n")
            print("Admin Details")
            for i in self.Admin_details:
                print(i , ":" , self.Admin_details[i])
        except Exception as e:
            print("\nPlease provide valid info")
    def Admin_login(self):
        try:
            print("Welcome to {} Restaurant \n".format(self.Restro_Name))
            Email = input("Enter mail id : ")
            Password = input("Enter Passwrod : ")
            if len(self.Admin_details.values()) != 0:
                if Email == self.Admin_email and Password == self.Admin_password:
                    print("Admin Login Successfull")
                    while True:
                        print("Select your Option")
                        print("1. Add food item\n2. Edit Food items\n3. View Food items\n4. Del Food items\n5. Exit")
                        z = input("Enter your Option : ")
                        if z == "1":
                            self.Add_food_item()
                        elif z == "2":
                            self.Edit_food_item()
                        elif z == "3":
                            self.View_food_item()
                        elif z == "4":
                            self.Del_food_item()
                        elif z == "5":
                            break 
                        else:
                            print("Plese Enter Valid Option")
                else:
                    print("Please Enter valid mail or Password")
            else:
                print("Admin Not found")
        except Exception as e :
            print("\n Provide Valid Inputs")
    def Add_food_item(self):
        try:
            Name = input("Enter Food item name : ")
            quantity = float(input("Enter the quantity : "))
            price = int(input("Enter the Price : "))
            stock = int(input("Enter the Stock : "))
            discount = int(input("Enter the discount : "))
            food_item = {"Name" : Name , "Quantity" : quantity , "Price" : price , "Stock" : stock , "Discount" : discount}
            self.food_id = len(self.food)+1
            self.food[self.food_id] = food_item
            print("Food item added successfully")
            print("\n Newly added food items : " , self.food , " \n ")
        except Exception as e:
            print("\nPlease provide valid inputs")
    def Edit_food_item(self):
        try:
            x = int(input("Enter food id that you want to update :  "))
            if x in self.food.keys():
                print("1. Update food name\n2. update quantity\n3. update price\n4. update stock\n5. update discount")
                y = input("Enter the key that you want edit : ")
                if y == "1":
                    self.food[x]["Name"] = input("Enter your new food name : ")
                    print("\n Successfully Updated \n")
                elif y == "2":
                    self.food[x]["Quantity"] = float(input("Enter updated quantity : "))
                    print("\n Successfully Updated \n")
                elif y == "3":
                    self.food[x]["Price"] = int(input("Enter updated price : "))
                    print("\n Successfully Updated \n")
                elif y == "4":
                    self.food[x]["Stock"] = int(input("Enter the updated stock : "))
                    print("\n Successfully Updated \n")
                elif y == "5":
                    self.food[x]["Discount"] = int(input("Enter updated discount : "))
                    print("\n Successfully Updated \n")
                else:
                    print("Invalid option")
            else:
                print("Incorrect food id")
        except Exception as e:
            print("Please provide valid inputs")
    def View_food_item(self):
        print("list of food items")
        if len(self.food) != 0:
            for i in self.food:
                print("food_id: " , {i})
                for j in self.food[i]:
                    print(j , ":" ,self.food[i][j])
                print()
        else:
            print("Food items not found")
    def Del_food_item(self):
        try:
            print("Enter the food id that you want to delete : ")
            x = int(input("Enter the food id : "))
            if x in self.food.keys():
                del self.food[x]
                print("Food item deleted successfully")
                print("The remaining food items are : " , self.food)
            else:
                print("Incorrect food id")
        except Exception as e:
            print("Please provide valid input")
    def user_register(self):
        try:
            print("Welcome to {} Restaurant".format(self.Restro_Name))
            self.fullname = input("Enter a fullname :  ")
            self.mobile_number = int(input("Enter your number : "))
            self.mail_id = input("Enter your mail id : ")
            self.password = input("Enter your password : ")
            self.address = input("Enter your Address : ")
            self.user_details = {"Full-Name" : self.fullname , "Mobile Number" : self.mobile_number , "Mail-id" : self.mail_id , "Password" : self.password , "Address" : self.address}
            print("\n User Account Created Successfully\n")
            print("User Details")
            for i in self.user_details:
                print(i , ":" , self.user_details[i])
        except Exception as e:
            print("\n Enter Valid inputs")
    def user_login(self):
        try:
            print("Welcome to {} Restaurant \n".format(self.Restro_Name))
            mail_id = input("Enter mail id : ")
            password = input("Enter Passwrod : ")
            if len(self.user_details.values()) != 0:
                if mail_id == self.user_details["Mail-id"] and password == self.user_details["Password"]:
                    print("User Login Successfull")
                    while True:
                        print("1. Place New Order\n2. Check Order History\n3. Update Your Profile\n4.Exit")
                        a = input("Select Your Option : ")
                        if a == "1":
                            self.Place_Order()
                        elif a == "2":
                            self.ORDERED_HISTORY()
                        elif a == "3":
                            self.Update_Details()
                        elif a == "4":
                            break
                        else:
                            print("Enter Valid option")
                else:
                    print("Incorrect Mail-id or Password")
            else:
                print("There is no acount with this Mail ID")
        except exception as e:
            print("\nInvalid Inputs")
    def Place_Order(self):
        try:
            if len(self.food) != 0:
                print("Menu Card")
                menu = []
                for items in self.food:
                    menu.append([self.food[items]["Name"],self.food[items]["Quantity"],self.food[items]["Price"]])
                for i in range(len(menu)):
                    print(i + 1, ":" , menu[i])
                while True:
                    print("Select your option")
                    print("1. Place Order\n2. Exit" )
                    b = input("Enter your option : ")
                    if b == "1":
                        print("Select your Food Items from our Menu")
                        c = input().split(",")
                        for i in c:
                            d = int(i)
                            if d <= len(menu):
                                self.ordered_item.append(menu[d-1])
                            else:
                                print("Out Of Stock")
                        print("Your Selected items are : ")
                        for j in self.ordered_item:
                            print(j)
                    elif b == "2":
                        break
                    else:
                        print("Please Select a valid option")
            else:
                print("There is no food")
        except eception as e:
            print("Please Provide Valid inputs")
    def ORDERED_HISTORY(self):
        for i in self.ordered_item:
            print(i)
    def Update_Details(self):
        try:
            while True:
                print("1. Edit Full Nmae\n2. Edit Mobile Number\n3. Edit Mail Id\n4. Change Password\n5. Edit Address\n6. Exit")
                u = input("Select Your Option :  ")
                if u == "1":
                    self.user_details["Full-Name"] = input("Enter your Updated Full Name :  ")
                    print("Your Full Name is Updated Successfully")
                elif u == "2":
                    self.user_details["Mobile Number"] = input("Enter Your Updated Mobile Number :  ")
                    print("Your Mobile Number is Updated Successfully")
                elif u == "3":
                    self.user_details["Mail-id"] = input("Enter Your Updated Mail-id :  ")
                    print("Your Mail-id is updated successfully")
                elif u == "4":
                    self.user_details["Password"] = input("Enter your new password :  ")
                elif u == "5":
                    self.user_details["Address"] = input("Enter your Updated Address :  ")
                    print("Your Address Updated Successfully")
                elif u == "6":
                    break
                else:
                    print("Select valid option")
                if u in ["1","2","3","4","5"]:
                    for i in self.user_details:
                        print(i , ":" , self.user_details[i])
                else:
                    print("You Entered Wrong Data")
        except exception as e:
            print("Invalid Inputs")
try:
    def main():
        anudeep = Restaurant()
        print("\n WELCOME TO THE PARK RESTAURANT")
        while True:
            print("1. Admin\n2. User\n3. Exit")
            k = input("Select Your Option :  ")
            if k == "1":
                while True:
                    print("Select the Below Option")
                    print("1. Admin_to_Register\n2. Admin_Login\n3. Exit")
                    l = input("Enter your option : ")
                    if l == "1":
                        anudeep.Admin_register()
                    elif l == "2":
                        anudeep.Admin_login()
                    elif l == "3":
                        break
                    else:
                        print("Please select valid input")
            elif k == "2":
                while True:
                    print("1. User Register\n2. User Login\n3. Exit")
                    m = input("Select Your Option :  ")
                    if m == "1":
                        anudeep.user_register()
                    elif m == "2":
                        anudeep.user_login()
                    elif m == "3":
                        break
                    else:
                        print("Please Select Your Valid Option")
            elif k == "3":
                break
            else:
                print("OOPS! You Selected Wrong Option PLease Select Valid option")
except exception as e:
    print("PLEASE SELECT A VALID OPTION")
if __name__ == '__main__':
    main()
print("*****************************************************************************************************************")
print("THANK YOU FOR VISITING OUR RESTAURANT")
print("*****************************************************************************************************************")
```
